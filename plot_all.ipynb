```python
df = pd.read_csv('kids_vsl_tbt_famphase.csv')
groups = df.groupby('id')
n = len(groups)
print n
```

```python
fig, axs = plt.subplots(8,6, figsize=(22, 20),)
fig.subplots_adjust(hspace = .3, wspace=.3)

axs = axs.ravel()

for idx, (name, group) in enumerate(groups):
    axs[idx].scatter(group.trial_mask, group.rt)
    axs[idx].set_title(name)
    
plt.savefig('plot_SSR_33TDv33Adult/vsl_tbt_kids_allpar_scatter.png', bbox_inchex='right')
```
